```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Flatten, Dense

(x_train, y_train),(x_test, y_test) = tf.keras.datasets.mnist.load_data()

x_train = x_train/255
x_test = x_test/255

INPUT_SHAPE = 28 * 28
```

```
model = tf.keras.models.Sequential()

model.add(Flatten(input_shape=(28,28)))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=10, activation='softmax'))

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.summary()
```

```
callback = model.fit(x=x_train, y=y_train, epochs=10, validation_data=(x_test, y_test), validation_steps=1)
```

```
plt.plot(callback.epoch, callback.history['val_accuracy'], label='val_acc')
plt.plot(callback.epoch, callback.history['accuracy'], label='acc')
plt.show()
```

```
plt.plot(callback.epoch, callback.history['val_loss'], label='val_loss')
plt.plot(callback.epoch, callback.history['loss'], label='loss')
plt.show()
```

```
predictions = model.predict(x_test[:15])

for i, prediction in enumerate(predictions):
  plt.imshow(x_test[i])
  plt.show()

  print(prediction.argmax())
```
